**Task 4: Data Cleaning**

```python
data.isnull().sum()
```

Handling Null Values

```python
data['neighbourhood'] = data['neighbourhood'].fillna('Unknown')
```

```python
data['latitude'] = data['latitude'].ffill()
```

```python
data['longitude'] = data['longitude'].ffill()
```

```python
data['room_type'] = data['room_type'].ffill()
```

Filling Median Values in Place Of null Values 

```python
data['price'] = data['price'].fillna(data['price'].median())
```

Filling mean in place of null values

```python
data['minimum_nights'] = data['minimum_nights'].fillna(data['minimum_nights'].mean())
```

Filling mean in place of null values

```python
data['number_of_reviews'] = data['number_of_reviews'].fillna(data['number_of_reviews'].mean())
```

Changing String To DateTime Format

```python
data['last_review'] = pd.to_datetime(data['last_review'])
```

```python
data['last_review'] = data['last_review'].ffill()
```

```python
data['reviews_per_month'] = data['reviews_per_month'].fillna(data['reviews_per_month'].mean())
```

```python
data['calculated_host_listings_count'] = data['calculated_host_listings_count'].fillna(data['calculated_host_listings_count'].median())
```

```python
data['availability_365'] = data['availability_365'].fillna(data['availability_365'].mean())
```

```python
data['number_of_reviews_ltm'] = data['number_of_reviews_ltm'].fillna(data['number_of_reviews_ltm'].mean())
```

```python
data.isnull().sum()
```

```python
# type casting 
# changing data types 
data['id'] = data['id'].astype(object)

data['host_id'] = data['host_id'].astype(object)
data.dtypes
```

```python
# dealing with duplicates rows
data.duplicated().sum()
```

```python
duplicates = data[data.duplicated()]
duplicates
```

```python
# id duplicates
id_duplicates= data[data.duplicated(subset = ['id'])]
id_duplicates
```

```python
# droping duplicate values
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

#### EDA
**Task 5: Data Analysis**

**Univariate Analysis**

```python
# idenfying outliers in price
df = data[data['price']<1500]

sns.boxplot(data=df, x='price')
plt.show()
```

```python
#Price distribuion
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='price', bins=100)
plt.title('Price Distribuition')
plt.ylabel("Frequency")
plt.show()
```

```python
#Price distribuion
plt.figure(figsize=(6, 3))
sns.histplot(data=df, x='availability_365')
plt.title('availability_365 Distribuition')
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

**Feature Engineering**

```python
# ['price per bed']

df['price per bed']= df['price']/df['beds']
df.head()
```

```python
# average price per bed
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

**Bi Variable Analysis**
One variable depenency in another variable

```python
# price dependency on neighbourhood
sns.barplot(data=df,x='neighbourhood_group',y='price',hue='room_type')
plt.show()
```

```python
# number of reviews and price rel
plt.figure(figsize=(8, 5))
plt.title("Locality and Review Dependency")
sns.scatterplot(data=df,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.show()
```

```python
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
plt.show()
```

```python
#Geographical Distribution of AirBnb Listing
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.title("Geographical Distribution of AirBnb Listing")
plt.show()
```

```python
# heat map - correlation of one variable with others for numerical column

corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True)
plt.show()
```

```python
sns.boxplot(x='room_type', y='price', data=df)
plt.title("Price by Room Type")
plt.show()
```

```python
sns.histplot(df['calculated_host_listings_count'], bins=30)
plt.title("Host Listings Count")
plt.show()
```

```python
df['host_name'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Hosts by Listings")
plt.show()
```

```python
sns.boxplot(x='bedrooms', y='price', data=df)
plt.title("Bedrooms vs Price")
plt.show()
```

```python
df['room_type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Room Type Distribution")
plt.show()
```

```python
df.groupby(df['last_review'].dt.year)['number_of_reviews'].sum().plot()
plt.title("Reviews Over Time")
plt.show()
```
